```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# Create engine using the `hawaii.sqlite` database file
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()

# Use the Base class to reflect the database tables
Base.prepare(autoload_with=engine)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Assign the measurement class to a variable called `Measurement` and
# assign the station class to a variable called `Station`
Measurement = Base.classes.measurement
Station = Base.classes.measurement
```

```python
# Create a session
session =  Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results.
# Starting from the most recent data point in the database.
recent_point = session.query(Measurement.date).order_by(Measurement.date.desc()).first()

# Calculate the date one year from the last date in data set.
last_twelve_months = dt.datetime(2017, 8, 23) - dt.timedelta(days=365)
one_year_ago = dt.date(last_twelve_months.year, last_twelve_months.month, last_twelve_months.day)

# Perform a query to retrieve the data and precipitation scores
dp_scores = session.query(Measurement.date, Measurement.prcp).order_by(Measurement.date.asc()).\
        filter(Measurement.date>=one_year_ago).all()

dp_scores

# Save the query results as a Pandas DataFrame. Explicitly set the column names
data_df =pd.DataFrame(dp_scores, columns=['date', 'precipitation'])

# Sort the dataframe by date
data_df.sort_values(by=['date'], ascending=True)

#drop nulls
clean_df = data_df.dropna()
clean_df

# Use Pandas Plotting with Matplotlib to plot the data
clean_df.plot('date', 'precipitation')

plt.xlabel("Date",  size = 12)
plt.ylabel("mm", size = 12)
plt.title("Hawaii Precipitation from Aug 2016 - Aug 2017", size = 12)
plt.legend(["precipitation"])
plt.xticks(rotation=90, size = 8)
plt.yticks(size = 8)
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
clean_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
session.query(Station.id).count()
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
active_stations = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station)\
.order_by(func.count(Measurement.station).desc()).all()
active_stations
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
# most active station is top one 2772
session.query(func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs))\
    .filter(Measurement.station=='USC00519281').all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
year_temp = session.query(Measurement.tobs).filter(Measurement.station=='USC00519281')\
    .filter(Measurement.date>='2016-08-23').all()

#Plot results
results_temp=pd.DataFrame(year_temp,columns=['tobs'])
results_temp.plot.hist(bins=12)
plt.tight_layout()
plt.title("Temperature in Hawaii Stations for Aug 2016 - Aug 2017", size=13)
plt.xlabel("Temperature")
plt.ylabel("Frequency")

plt.show()
```

# Close Session

```python
# Close Session
session.close()
```
